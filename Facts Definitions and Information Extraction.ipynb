```python
import spacy
from spacy import displacy
from itertools import combinations
nlp = spacy.load('en_core_web_lg')

# load NeuralCoref and add it to the pipe of SpaCy's model
import neuralcoref
coref = neuralcoref.NeuralCoref(nlp.vocab,conv_dict={'Singapore': ['country','region','area']})
nlp.add_pipe(coref, name='neuralcoref')

# You're done. You can now use NeuralCoref the same way you usually manipulate a SpaCy document and it's annotations.
file_name = 'corpus.txt'
text = open(file_name,  encoding='utf-8').read()
#text="""
#Singapore, officially the Republic of Singapore, is a sovereign island city-state in maritime Southeast Asia. It lies about one degree of latitude (137 kilometres or 85 miles) north of the equator, off the southern tip of the Malay Peninsula, bordering the Straits of Malacca to the west, the Riau Islands (Indonesia) to the south, and the South China Sea to the east. """
doc = nlp(text)
doc = nlp(doc._.coref_resolved)
sents = list(doc.sents)
```

```python
from openie import StanfordOpenIE
kg = []

client = StanfordOpenIE()
for sent in sents:
    text = sent.text
    print('Text: %s' % text)
    #print(client.annotate(text))
    sent_vector =[]
    for token in sent:
        sent_vector.append(token.vector_norm)
    for triple in client.annotate(text):
        #print(triple)
        #print(triple["subject"])
        #print(triple["relation"])
        #print(triple["object"])       
        kg.append([triple["subject"],triple["relation"],triple["object"],sent.text ,sent_vector])
```

```python
import pandas as pd  

df = pd.DataFrame(kg, columns =['subject', 'predicate','object','original sent','original sent vector']) 
df.to_csv("fact_base_full.csv") 
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx_viewer import Viewer
import pydot

edges = []
edge_labels = {}
for kg_item in kg:
    #print(kg_item)
    edges.append([kg_item[0],kg_item[2]])
    #edges.append([kg_item[1],kg_item[2]])
    edge_labels[(kg_item[0],kg_item[2])] = kg_item[1]
    



G = nx.Graph()
G.add_edges_from(edges)

for kg_item in kg:
    #print(kg_item)
    edges.append([kg_item[0],kg_item[2]])
    #edges.append([kg_item[1],kg_item[2]])
    edge_labels[(kg_item[0],kg_item[2])] = kg_item[1]
    G.edges[kg_item[0],kg_item[2]]['label'] = kg_item[1]


pos = nx.spring_layout(G)
plt.figure()    
nx.draw(G,pos,edge_color='black',width=1,linewidths=1,\
node_size=500,node_color='pink',alpha=0.9,\
labels={node:node for node in G.nodes()})
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)


plt.axis('off')
plt.show()

nx.write_graphml(G, "gmltest.graphml")
```
